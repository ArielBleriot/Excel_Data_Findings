```python
import pandas as pd
df = pd.read_csv('retail-orders/orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
df.head(20)
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(10)
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']= df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
df.head(5)
```

```python
import sqlalchemy as sal
from sqlalchemy.sql import text  

# Connection string (Windows Authentication)
engine = sal.create_engine(
    'mssql+pyodbc://@localhost\\SQLEXPRESS/master?'
    'driver=ODBC+Driver+17+for+SQL+Server&'
    'Trusted_Connection=yes&'
    'Encrypt=yes&'
    'TrustServerCertificate=yes'
)

try:
    conn = engine.connect()
    print("Successfully connected to SQL Server!")
    
    # Corrected query execution with text()
    result = conn.execute(text("SELECT @@VERSION")).scalar()
    print(f"SQL Server Version:\n{result}")
    
    conn.close()
except Exception as e:
    print(f"Connection failed: {e}")
```

```python
df.to_sql(
    name='orders',  
    con=engine,    
    index=False,
    if_exists='append'
)
```
